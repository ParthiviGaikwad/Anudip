### **Compare the distribution of key variables**

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from scipy.stats import ttest_ind
# Convert categorical variables to numeric (0 and 1) if needed
df['HighBP'] = df['HighBP'].astype(int)
df['HighChol'] = df['HighChol'].astype(int)
df['Smoker'] = df['Smoker'].astype(int)
df['Stroke'] = df['Stroke'].astype(int)

# Mean values for categorical variables grouped by HeartDiseaseorAttack
for var in ['HighBP', 'HighChol', 'Smoker', 'Stroke']:
    print(f"\nMean of {var} grouped by HeartDiseaseorAttack:")
    print(df.groupby('HeartDiseaseorAttack')[var].mean())

# Box plot for BMI
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df)
plt.title('BMI Distribution by Heart Disease')
plt.show()

# Chi-Square Tests for categorical variables
for var in ['HighBP', 'HighChol', 'Smoker', 'Stroke']:
    contingency_table = pd.crosstab(df['HeartDiseaseorAttack'], df[var])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    print(f"Chi-Square test for {var}: p-value = {p}")

# T-Test for BMI
bmi_heart_disease = df[df['HeartDiseaseorAttack'] == 1]['BMI']
bmi_no_heart_disease = df[df['HeartDiseaseorAttack'] == 0]['BMI']

if bmi_heart_disease.shape[0] > 30 and bmi_no_heart_disease.shape[0] > 30:
    t_stat, p_value = ttest_ind(bmi_heart_disease, bmi_no_heart_disease, equal_var=False)
else:
    from scipy.stats import mannwhitneyu
    t_stat, p_value = mannwhitneyu(bmi_heart_disease, bmi_no_heart_disease)

print(f"Test for BMI: p-value = {p_value}")
print("""
Conclusion:
1. High Blood Pressure:
    (1) High blood pressure is significantly more prevalent among individuals with heart disease (75.03%) compared to those without (39.56%).
    (2) The Chi-square test indicates a strong association between high blood pressure and heart disease, meaning individuals with high blood pressure are at a much higher risk of developing heart disease.

2. High Cholesterol
  (1) High cholesterol is also much more prevalent in individuals with heart disease compared to those without.
  (2)The strong statistical association (p-value = 0.0) suggests that high cholesterol is a significant risk factor for heart disease.

3. Smoking:
  (1) Smoking is notably more common among individuals with heart disease than those without.
  (2) The association is statistically significant, reinforcing that smoking is a major risk factor for heart disease.

4. Stroke
  (1) Stroke is strongly associated with heart disease.
  (2) Individuals who have suffered a stroke are at a greater risk for heart disease.

5. Body Mass Index
    (1) The extremely low p-value for BMI suggests a very strong relationship between BMI and heart disease.
    (2) This implies that BMI is a significant risk factor, and individuals with higher BMI are more likely to suffer from heart disease.
""")
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# List of key variables to explore (including Diabetes)
key_variables = ['HighBP', 'HighChol', 'BMI', 'Smoker', 'Stroke', 'Diabetes']

# Loop through key variables and plot them separately for heart disease groups
for var in key_variables:
    # Print the numerical summaries
    print(f"\nSummary Statistics for {var}:")

    # Summary for individuals *with* heart disease
    if df[var].dtype == 'int64' or df[var].nunique() == 2:  # Categorical (0 or 1)
        heart_disease_counts = df[df['HeartDiseaseorAttack'] == 1][var].value_counts()
        print(f"Individuals with Heart Disease:\n{heart_disease_counts}")
    else:  # Continuous (like BMI)
        heart_disease_mean = df[df['HeartDiseaseorAttack'] == 1][var].mean()
        heart_disease_median = df[df['HeartDiseaseorAttack'] == 1][var].median()
        print(f"Individuals with Heart Disease - Mean: {heart_disease_mean}, Median: {heart_disease_median}")

    # Summary for individuals *without* heart disease
    if df[var].dtype == 'int64' or df[var].nunique() == 2:  # Categorical (0 or 1)
        no_heart_disease_counts = df[df['HeartDiseaseorAttack'] == 0][var].value_counts()
        print(f"Individuals without Heart Disease:\n{no_heart_disease_counts}")
    else:  # Continuous (like BMI)
        no_heart_disease_mean = df[df['HeartDiseaseorAttack'] == 0][var].mean()
        no_heart_disease_median = df[df['HeartDiseaseorAttack'] == 0][var].median()
        print(f"Individuals without Heart Disease - Mean: {no_heart_disease_mean}, Median: {no_heart_disease_median}")

    # Plot the distributions
    plt.figure(figsize=(18, 6))

    # Plot for individuals *with* heart disease
    plt.subplot(1, 2, 1)
    if df[var].dtype == 'int64' or df[var].nunique() == 2:  # Categorical (0 or 1)
        sns.countplot(data=df[df['HeartDiseaseorAttack'] == 1], x=var, color='orange')
        plt.title(f'Distribution of {var} for Individuals with Heart Disease')
    else:  # Continuous (like BMI)
        sns.kdeplot(data=df[df['HeartDiseaseorAttack'] == 1], x=var, fill=True, color='orange')
        plt.title(f'Distribution of {var} for Individuals with Heart Disease')

    plt.xlabel(var)
    plt.ylabel('Density' if df[var].dtype != 'int64' else 'Count')

    # Plot for individuals *without* heart disease
    plt.subplot(1, 2, 2)
    if df[var].dtype == 'int64' or df[var].nunique() == 2:  # Categorical (0 or 1)
        sns.countplot(data=df[df['HeartDiseaseorAttack'] == 0], x=var, color='blue')
        plt.title(f'Distribution of {var} for Individuals without Heart Disease')
    else:  # Continuous (like BMI)
        sns.kdeplot(data=df[df['HeartDiseaseorAttack'] == 0], x=var, fill=True, color='blue')
        plt.title(f'Distribution of {var} for Individuals without Heart Disease')

    plt.xlabel(var)
    plt.ylabel('Density' if df[var].dtype != 'int64' else 'Count')

    # Adjust layout and display
    plt.tight_layout()
    plt.show()

print("\n\n")
print("""
Conclusion:

1. High Blood Pressure: A significantly higher proportion of individuals with heart disease suffer from high blood pressure, suggesting it is a strong risk factor for heart disease.
2. High Cholesterol: High cholesterol is more prevalent among those with heart disease, highlighting it as another important risk factor.
3. Body Mass Index: While the distribution of BMI is similar between the two groups, slightly higher BMI values may indicate an increased risk of heart disease.
4. Smoking: Smoking is much more common among individuals with heart disease, reinforcing its known association with cardiovascular risk.
5. Stroke: Stroke is strongly associated with heart disease, indicating a significant overlap between the two conditions.
6. Diabetes: Diabetes, especially type 2, is more common in individuals with heart disease, reinforcing its role as a risk factor.
""")
```

```
import pandas as pd
import matplotlib.pyplot as plt


df['BMI'] = pd.to_numeric(df['BMI'], errors='coerce')
df['HeartDiseaseorAttack'] = df['HeartDiseaseorAttack'].astype(int)

def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['BMI'].apply(bmi_category)

# Create pivot table
pivot = pd.pivot_table(df, values='HeartDiseaseorAttack',
                       index='BMI_Category',
                       aggfunc='mean')

# Sort by BMI category
pivot = pivot.reindex(['Underweight', 'Normal weight', 'Overweight', 'Obese'])

# Create bar chart
pivot.plot(kind='bar', figsize=(10, 6))
plt.title('Heart Disease Prevalence by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Heart Disease Prevalence')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(pivot)
```
