```python
#finding any relation between MetaScore and IMDb rating

import numpy as np
import matplotlib.pyplot as plt

print("Relationship between IMDb Rating and MetaScore")
samp3=samp
#convert IMDb Rating into scale of 100 by multiplying by 10
samp3['IMDb Rating'] = samp['IMDb Rating'] * 10

X = samp3['Title']
Y_imdb = samp3['IMDb Rating']
Z_metascore = samp3['MetaScore']

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Y_imdb, 0.4, label='IMDb Rating')
plt.bar(X_axis + 0.2, Z_metascore, 0.4, label='MetaScore')

plt.xticks(X_axis, X, rotation=45, ha='right')
plt.xlabel("Movies")
plt.ylabel("Scores")
plt.title("IMDb Rating and MetaScore Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("IMDb ratings reflect user opinions, while MetaScores represent critic consensus.\nIMDb ratings can be influenced by fan bias, while MetaScores are based on professional critiques")

print("\n")
samp3['MetaScore Rank'] = samp3['MetaScore'].rank(ascending=False, method='min')
samp3['IMDb Rating Rank'] = samp3['IMDb Rating'].rank(ascending=False, method='min')

ranked_samp_by_MetaScore = samp3.sort_values(by=['MetaScore Rank'])
ranked_samp_by_IMDb = samp3.sort_values(by=['IMDb Rating Rank'])

#print(ranked_samp[['Title', 'MetaScore', 'IMDb Rating']])
print(ranked_samp_by_MetaScore[['Title', 'MetaScore']])
print("\n")
print(ranked_samp_by_IMDb[['Title', 'IMDb Rating']])

print("\n")
# Plot ranked MetaScores
plt.figure(figsize=(10, 6))
plt.barh(ranked_samp_by_MetaScore['Title'], ranked_samp_by_MetaScore['MetaScore'], color='skyblue')
plt.title('Ranked by MetaScore')
plt.xlabel('MetaScore')
plt.ylabel('Title')
plt.show()

# Plot ranked IMDb Ratings
plt.figure(figsize=(10, 6))
plt.barh(ranked_samp_by_IMDb['Title'], ranked_samp_by_IMDb['IMDb Rating'], color='orange')
plt.title('Ranked by IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Title')
plt.show()
print("Conclusion: There is a difference in scores. Some movies can be seen in top 5 while some can be seen in lower. \nWhereas There are noticeable differences between MetaScore and IMDb Rating rankings but not major.")
```

```python
data['IMDb Rating'] = pd.to_numeric(data['IMDb Rating'], errors='coerce')

# Define the time periods
time_periods = {
    '2010-2015': (2010, 2015),
    '2016-2020': (2016, 2020),
    '2021-2024': (2021, 2024)
}

# Initialize a dictionary to store the results
genre_ratings = {}

# Calculate the average IMDb ratings for each period
for period, (start_year, end_year) in time_periods.items():
    period_data = data[(data['Year'] >= start_year) & (data['Year'] <= end_year)]
    average_ratings = period_data.groupby('Genre')['IMDb Rating'].mean()
    genre_ratings[period] = average_ratings

# Convert the dictionary to a DataFrame for easier plotting
ratings_df = pd.DataFrame(genre_ratings)

# Plot the joint bar graph
ratings_df.plot(kind='bar', figsize=(12, 8))
plt.title('Average IMDb Ratings by Genre over Time Periods')
plt.xlabel('Genre')
plt.ylabel('Average IMDb Rating')
plt.xticks(rotation=45)
plt.legend(title='Time Periods')
plt.show()

print("\nCONCLUSION: In 2010-2015, no FANSTASY and THRILLER are released.\nIn 2016-2020,no BIOGRAPHY and THRILLER are released.\nBut in 2021-2024, all genres are released\nCOMEDY has been consistently liked.")
```
